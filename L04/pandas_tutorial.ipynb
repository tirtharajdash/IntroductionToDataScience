```python
data.drop('sepal_length_norm', axis=1, inplace=True)
data.head()
```

```python
data = data.append({'sepal_length': 10, 'sepal_width': 10, 'petal_length': 10}, ignore_index=True)
data.tail()
```

```python
#Drop this new row: data.index will return the row number. We know that we had inserted the data at the end
data.drop(max(data.index), axis=0, inplace=True)
data.tail()
```

```python
#insert again
data = data.append({'sepal_length': 10, 'sepal_width': 10, 'petal_length': 10}, ignore_index=True)
data.tail()
```

```python
#filling row with missing values
data.fillna(value={'petal_width': 0}, inplace=True)
data.tail()
```

```python
data.fillna(value={'Class': 'Iris-New'}, inplace=True)
data.tail()
```

```python
data.fillna(0)
data.tail()
```

#### sorting

```python
data.head()
```

```python
data.sort_values(by='sepal_length', ascending=False, inplace=True)
data.head()
```

Note that the inplace keyword means that the DataFrame will be overwritten, and hence no new DataFrame is returned. If instead of ascending = False we use ascending = True, the values are sorted in ascending order (i.e. from the smallest to the largest values).

If we want to return to the original order, we can sort by an index using the sort_index function and specifying axis=0:

```python
data.sort_index(axis=0, ascending=True, inplace=True)
data.head()
```
